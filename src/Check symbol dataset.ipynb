```python
import numpy, pickle
import matplotlib.pyplot as plt
```

```python
with open('../dataset/data.pickle', 'rb') as f:
        data = pickle.load(f)
```

```python
m = 10     # rows
n = 14     # columns
o = 210  # offset
fig, ax = plt.subplots(m, n, sharex=False, sharey=False, figsize=(n*2,m*2))
for i in range(m):
    for j in range(n):
        ax[i,j].imshow(data['img'][i*n + j + o])
        ax[i,j].set_title(data['label'][i*n + j + o])
        ax[i,j].set_xticks([])
        ax[i,j].set_yticks([])
```

```python
data2 = {'label': [], 'img': []}
for i, j in zip(data['label'], data['img']):
    if i == ',':
        data2['label'].append(i)
        data2['img'].append(j)

m = 35     # rows
n = 20     # columns
o = 0  # offset
fig, ax = plt.subplots(m, n, sharex=False, sharey=False, figsize=(n*2,m*2))
for i in range(m):
    for j in range(n):
        ax[i,j].imshow(data2['img'][i*n + j + o])
        ax[i,j].set_title(data2['label'][i*n + j + o])
        ax[i,j].set_xticks([])
        ax[i,j].set_yticks([])
```
